# Poids de l'√©vidence (PdE)

Nous calculons le poids de la preuve dans le but de regrouper des cat√©gories PdE similaires.

Pour ce faire, nous commencerons par estimer le poids de la preuve (PDE)
PdE = ln(%Bons/%Mauvais).

Le poids de la preuve indique la puissance pr√©dictive d'une variable ind√©pendante (home_ownership) par rapport √† la variable d√©pendante (bons_mauvais).

PdE est d√©crit comme une mesure de la s√©paration entre les bons et les mauvais clients.

"Clients mauvais" se r√©f√®re aux clients qui ont fait d√©faut sur un pr√™t,
et "Bons clients" se r√©f√®re aux clients qui ont rembours√© le pr√™t.

Tout cela est d√ª au fait que les cat√©gories avec un PdE similaire ont presque la m√™me proportion de bons et de mauvais. En d'autres termes, le comportement des deux cat√©gories est le m√™me.

# Valeur de la Information (VI)

L'analyse de la valeur d'information est une technique d'exploration de donn√©es
qui aide √† d√©terminer quelles variables d'un ensemble de donn√©es ont un pouvoir pr√©dictif ou une influence sur la valeur d'une variable d√©pendante sp√©cifique.

Notons que nous n'√©valuons pas le poids des cat√©gories au sein de la variable,
mais plut√¥t de la variable elle-m√™me.

VI = Somme (PdE * (proportion des bons dans la cat√©gorie -
proportion des mauvais dans la cat√©gorie))

```python
df1['VI'] = df1['PdE']*(df1['prop_n_buenos'] - df1['prop_n_malos'])
df1['sumatoriaVI'] = df1['VI'].sum()
```

# Automatizaci√≥n du Preprocessus des Variables Cat√©goriques

D√©finissons la fonction PdE_categorica qui automatisera le processus pr√©c√©dent pour calculer le PdE et VI pour toute variable d√©pendante (cat√©gorique) et ind√©pendante.

```python
def PdE_categorica(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    df = df.sort_values(['PdE'])
    df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categor√≠as
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
    
    
```

# Variables Continues

```python
# Nous copions et collons la fonction PdE_discrete
# Nous commentons les lignes de code o√π nous triions les valeurs par PdE et r√©initialisions les indices
```

```python
def PdE_continua(df, var_categorica, df_var_dependiente):
    df = pd.concat([df[var_categorica], df_var_dependiente], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                 df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0,1,3]]
    df.columns = [df.columns.values[0], 'observaciones', 'prop_bueno']
    df['proporcion_obs'] = df['observaciones']/df['observaciones'].sum()
    df['n_buenos'] = df['prop_bueno']*df['observaciones']
    df['n_malos'] = (1- df['prop_bueno'])*df['observaciones'] 
    df['prop_n_buenos'] = df['n_buenos']/df['n_buenos'].sum()
    df['prop_n_malos'] = df['n_malos']/df['n_malos'].sum() 
    df['PdE'] = np.log(df['prop_n_buenos']/df['prop_n_malos'])
    # df = df.sort_values(['PdE'])
    # df = df.reset_index(drop = True)
    df['delta_prop_n_buenos'] = df['prop_n_buenos'].diff().abs() 
    df['delta_PdE'] = df['PdE'].diff().abs() #Diferencia Absoluta entre categor√≠as
    df['VI'] = (df['prop_n_buenos'] - df['prop_n_malos']) * df['PdE']
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df['sumatoriaVI'] = df['VI'].sum()
    return df
```

Comme le r√©sultat sera un dataframe, nous n'avons pas besoin d'apporter de changements √† notre fonction de graphique, la seule diff√©rence sera que le graphique ne sera pas tri√© du PdE le plus bas au plus √©lev√©
